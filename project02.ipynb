```python
import pandas as pd

df = pd.read_csv("train.csv", encoding='latin1')
print(df.head())
print(df.columns)
```

```python
## 1. Data Loading and Preparation
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df = df.dropna(subset=['Order Date'])
df['Year'] = df['Order Date'].dt.year
```

```python
print(df[['Order Date', 'Year']].head())
```

```python
## 2. Sales Trend Over Time
```

```python
import matplotlib.pyplot as plt

sales_trend = df.groupby('Year')['Sales'].sum()

plt.figure()
sales_trend.plot(marker='o')
plt.title("Sales Over Time")
plt.show()
```

```python
# Insight:
# Sales decreased slightly from 2015 to 2016, then increased strongly in 2017 and remained high in 2018.
# This suggests an overall positive sales trend in later years.
```

```python
## 3. Sales by Category
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure()
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
# Insight:
# Technology generates the highest sales among all categories, followed by Furniture, while Office Supplies has the lowest sales.
# This suggests that Technology products are the main revenue driver.
```

```python
## 4. Sales by Region
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure()
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
# Insight:
# The West region generates the highest sales, followed by the East and Central regions, while the South region has the lowest sales.
# This suggests that the West region is the strongest market.
```

```python
## 5. Sales by Customer Segment
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure()
segment_sales.plot(kind='bar')
plt.title("Sales by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Sales")
plt.show()
```

```python
# Insight:
# The Consumer segment generates the highest sales, followed by the Corporate segment, while the Home Office segment has the lowest sales.
# This indicates that individual consumers are the primary source of revenue.
```

```python
## 6. Final Conclusion
```

```python
# The analysis shows that sales have increased over time, especially after 2016.
# Technology is the top-performing category, making it the main revenue driver.
# The West region leads in sales, indicating a strong market presence.
# The Consumer segment contributes the most to total sales, suggesting that individual customers are the key target group.
# Overall, the business performs best in technology products, western regions, and consumer markets.
```
